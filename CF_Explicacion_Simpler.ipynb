```python
import numpy as np
from sklearn.impute  import SimpleImputer # llamar libreria
```

```python
data = np.array([[49, np.nan, 7], [34,58,np.nan], [np.nan, 87,98]])
```

```python
data
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') #median, most_frequent # crear configura
imputar = imputer.fit(data) # promedio
data_imputada = imputar.transform(data) # llenar
```
