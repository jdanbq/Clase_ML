```python
def crear_trenes(rango):
  n_pasos = 3
  x_train = []
  y_train = []

  for index_inicial in np.arange(0,len(rango)-3,1):
    index_final = index_inicial + n_pasos

    x_secuencia = rango[index_inicial:index_final]
    y_secuencia = rango[index_final]

    #print(x_secuencia, '    ', y_secuencia)

    x_train.append(x_secuencia)
    y_train.append(y_secuencia)

  x_train = np.array(x_train)
  y_train = np.array(y_train)

  return(x_train,y_train)
```

```python
np.shape(x_train) #numero muesras, numero de pasos
```

### Lineal

```python
maquina_lineal = SVR(kernel='linear')
model_lineal = maquina_lineal.fit(x_train, y_train) # entrenamos
```

```python
x_pred = np.array([70,80,90])
```

```python
np.shape(x_pred)
```

```python
x_pred = x_pred.reshape(1,3)
```

```python
np.shape(x_pred)
```

```python
prediccion = model_lineal.predict(x_pred)
```

```python
prediccion
```

### **Kernel Poly**

```python
maquia_poly =  SVR(kernel='poly')
model_poly = maquia_poly.fit(x_train, y_train) # entrenamos
```

```python
prediccion_poly = model_poly.predict(x_pred)
prediccion_poly
```

### **Kernel RBF**

```python
maquia_rbf =  SVR(kernel='rbf', C = 10**10, gamma= 0.1)
model_rbf = maquia_rbf.fit(x_train, y_train) # entrenamos
```

```python
prediccion_rbf = model_rbf.predict(x_pred)
prediccion_rbf
```
